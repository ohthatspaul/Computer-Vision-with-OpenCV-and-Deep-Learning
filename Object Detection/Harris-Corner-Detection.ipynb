```python
'''
A corner can be interpreted as the junction of two edges, where an edge is a sudden
change in image brightness. 

A corner is a point whose local neighborhood stands in two dominant and different edge 
directions. 

There are various corner detection algorithms including:

1. Harris Corner Detection
2. Shi-Tomasi Corner Detection
'''
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#color chessboard
flat_chess_board = cv2.imread('../DATA/flat_chessboard.png')
flat_chess_board = cv2.cvtColor(flat_chess_board,cv2.COLOR_BGR2RGB)

#grayscale chessboard
gray_flat_chess = cv2.cvtColor(flat_chess_board, cv2.COLOR_BGR2GRAY)

#real chessboard color
real_chess_board = cv2.imread('../DATA/real_chessboard.jpg')
real_chess_board = cv2.cvtColor(real_chess_board, cv2.COLOR_BGR2RGB)

#grayscale real chessboard
gray_real_chess_board = cv2.cvtColor(real_chess_board, cv2.COLOR_BGR2GRAY)

```

```python
#convert grayscale image to floating point values
gray = np.float32(gray_flat_chess)

#apply the algorithm
destination = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)

#dilate result to mark out the corners for plotting the values
destination = cv2.dilate(destination,None)

'''
for visualization: wherever the result of the Harris Corner is greater than
one percent of the max value, reassign that to be red color. 
That is, mark it as a corner
'''
flat_chess_board[destination>0.01*destination.max()] = [255,0,0]
plt.imshow(flat_chess_board)
```

```python
#on the real chessboard
gray = np.float32(gray_real_chess_board)

#apply the algorithm
destination = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
#dilate result to mark out the corners for plotting the values
destination = cv2.dilate(destination,None)

'''
for visualization: wherever the result of the Harris Corner is greater than
one percent of the max value, reassign that to be red color. 
That is, mark it as a corner
'''
real_chess_board[destination>0.01*destination.max()] = [255,0,0]
plt.imshow(real_chess_board)
```
