```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

'''Contours are defined simply as a curve joining all the 
continous points(along the boundary), having the same color or intensity.
They are useful for shape analysis and object detection and recogntion'''

image = cv2.imread('../DATA/internal_external.png',0)
#extract internal and external contours
image_return, contours, hierarchy = cv2.findContours(image,cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

external_contours = np.zeros(image_return.shape)


'''this goes through the contours list and checks if the
hierachy for that corresponding contour is good and if it is, onto the
black external contours (np.zeros), draw on that specific contour and we will
be able to view it'''
#external
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours,contours,i,255,-1)
        
internal_contours = np.zeros(image_return.shape)
#internal
for i in range(len(contours)):
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contours,contours,i,255,-1)
        


plt.subplot(121)

plt.imshow(internal_contours, cmap='gray', aspect='auto')
plt.title('Internal Contours')
    
plt.subplot(122)
plt.imshow(external_contours, cmap='gray', aspect='auto')
plt.title('External Contours')
plt.show()
print('\n \n')
```
