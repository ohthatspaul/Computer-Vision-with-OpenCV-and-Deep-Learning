```python
#connecting to camera
import cv2

capture = cv2.VideoCapture(0)

width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

#save video to file
writer = cv2.VideoWriter('paulvideo.mp4',cv2.VideoWriter_fourcc(*'DIVX'),
                         20,(width,height))
#display the image
while True:
    ret,frame = capture.read()
    
    #Operations (Drawing)
    writer.write(frame)
    
    #convert fram into grayscale
   # gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    #directly stream the video
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
        
capture.release()
writer.release()
cv2.destroyAllWindows()
```

```python
#opening files
import cv2
import time

captured = cv2.VideoCapture('paulvideo.mp4')
if captured.isOpened() == False:
    print('ERROR FILE NOT FOUND OR WRONG CODEC USED!')
    
while captured.isOpened():
    ret, frame = captured.read()
    
    if ret == True:
        #writer 20 FPS
        time.sleep(1/20)
        cv2.imshow('frame',frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        
    else:
        break
captured.release()
cv2.destroyAllWindows()
```

```python
#drawing on video
import cv2

capture = cv2.VideoCapture(0)

width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

#top left corner
x= width//2
y=height//2

#width and height of Rectanlge

w= width//4
h = height//4

#bottom right x+w , y+h

while True:
    ret, frame = capture.read()
    #draw rectangle
    cv2.rectangle(frame,(x,y),(x+w,y+h),color = (0,0,255),thickness=5)
    
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
capture.release()
cv2.destroyAllWindows()
```

```python
#drawing on video
import cv2

#callback function to draw rectangle
def draw_rectangle(event,x,y,flags,param):
    global point1, point2, topLeft_clicked, bottomRight_clicked
    if event == cv2.EVENT_LBUTTONDOWN:
        
        #reset the rectangle if rectangle is here
        if topLeft_clicked == True and bottomRight_clicked == True:
            point1 = (0,0)
            point2 = (0,0)
            topLeft_clicked = False
            bottomRight_clicked = False
           
            
        if topLeft_clicked == False:
            point1 = (x,y)
            topLeft_clicked = True
            
        elif bottomRight_clicked == False:
            point2 = (x,y)
            bottomRight_clicked = True
            

#global variables

point1=(0,0) #top left
point2 = (0,0) #bottom right

topLeft_clicked = False
bottomRight_clicked = False

#capture video
capture = cv2.VideoCapture(0)
#connect to callback
cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_rectangle)


while True:
    ret, frame = capture.read()
    #draw on frame based off the global variables
    if topLeft_clicked:
        cv2.circle(frame,center = point1, radius = 3,
                   color = (0,0,255), thickness=-1)
        
    if topLeft_clicked and bottomRight_clicked:
        cv2.rectangle(frame,point1,point2,(0,0,255),3)
        
    cv2.imshow('Test',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
capture.release()
cv2.destroyAllWindows()
```
