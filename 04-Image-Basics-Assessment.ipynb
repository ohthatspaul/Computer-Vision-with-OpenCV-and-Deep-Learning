#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
img_triangle = cv2.cvtColor(img,cv2.COLOR_BGR2RGB).copy()

#triangle points
pt1 = (200,710)
pt2 = (410,380)
pt3 = (600,710)

cv2.circle(img_triangle,pt1,2,(255,0,0))
cv2.circle(img_triangle,pt2,2,(255,0,0))
cv2.circle(img_triangle,pt3,2,(255,0,0))

triangle_contour = np.array([pt1,pt2,pt3],np.int32)
cv2.drawContours(img_triangle,[triangle_contour],0,(255,0,0),thickness=20)
plt.imshow(img_triangle)



```
